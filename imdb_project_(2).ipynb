# مقدمه ای بر سیستم های توصیه **گر**

در دنیای امروزی، با رشد روزافزون اطلاعات و گسترش منابع مختلف داده، کاربران اغلب با حجم زیادی از انتخاب‌ها مواجه می‌شوند که تصمیم‌گیری را دشوار می‌کند. در این میان، سیستم‌های توصیه‌گر  به‌عنوان ابزاری هوشمند و کارآمد برای تسهیل انتخاب و شخصی‌سازی تجربه کاربران توسعه یافته‌اند. این سیستم‌ها به‌طور گسترده در پلتفرم‌های مختلف از جمله فروشگاه‌های آنلاین، سرویس‌های استریم موسیقی و فیلم، شبکه‌های اجتماعی و حتی سیستم‌های آموزشی استفاده می‌شوند.

هدف اصلی سیستم‌های توصیه‌گر، ارائه پیشنهادات مناسب به کاربران است، به‌گونه‌ای که این پیشنهادات با علایق، نیازها و رفتار گذشته کاربر همخوانی داشته باشد. به عبارت دیگر، این سیستم‌ها به کاربران کمک می‌کنند تا به سرعت محتوای مورد نظر خود را در میان حجم انبوهی از داده‌ها بیابند.

# سیستم های توصیه گر

   دو رویکرد اصلی در سیستم‌های توصیه‌گر هستند که برای پیشنهاد محصولات، خدمات یا محتوا به کاربران استفاده می‌شوند. در ادامه، توضیح مختصری برای هر کدام ارائه شده است:

### **Collaborative Filtering (فیلترینگ مشارکتی)**  
این روش بر اساس **رفتار کاربران مشابه** عمل می‌کند. ایده اصلی این است که اگر دو کاربر در گذشته انتخاب‌های مشابهی داشته باشند، احتمالاً در آینده نیز علاقه‌مندی‌های مشابهی خواهند داشت.  

#### ویژگی‌ها:  
- از اطلاعات تعامل کاربران (مانند امتیازدهی یا خریدها) استفاده می‌کند.  
- نیازی به اطلاعات خاصی درباره ویژگی‌های آیتم‌ها ندارد.  
- چالش اصلی آن در مواجهه با مشکل **شروع سرد** (Cold Start) برای کاربران یا آیتم‌های جدید است.  

#### مثال:  
اگر کاربر A و کاربر B هر دو به فیلم‌های مشابهی امتیاز بالا داده‌اند و کاربر A فیلم دیگری را دوست دارد که کاربر B هنوز ندیده است، این فیلم به کاربر B پیشنهاد می‌شود.  


### **Content-Based Filtering (فیلترینگ مبتنی بر محتوا)**  
این روش بر اساس **ویژگی‌های آیتم‌ها** و **ترجیحات کاربر** عمل می‌کند. سیستم به دنبال ویژگی‌های مشترک بین آیتم‌هایی است که کاربر قبلاً به آن‌ها علاقه نشان داده و سپس آیتم‌های مشابه را پیشنهاد می‌دهد.  

#### ویژگی‌ها:  
- به اطلاعات دقیق درباره ویژگی‌های آیتم‌ها نیاز دارد.  
- تمرکز بر شخصی‌سازی بر اساس تاریخچه کاربر است.  
- ممکن است آیتم‌های جدید خارج از علایق شناخته‌شده کاربر را پیشنهاد نکند.  

#### مثال:  
اگر کاربری کتاب‌هایی در ژانر علمی‌تخیلی مطالعه کرده است، سیستم کتاب‌های دیگری در همین ژانر را پیشنهاد می‌دهد.  


### مقایسه کلی:  
- **Collaborative Filtering**: مناسب برای کشف علایق جدید و ارائه پیشنهادات متنوع.  
- **Content-Based Filtering**: برای پیشنهادات دقیق بر اساس سلیقه کاربر مفید است.

```python
import pandas as pd
from math import sqrt
import numpy as np

# بارگذاری داده‌ها
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
```

```python
movies_df['year'] = movies_df['title'].str.extract(r'\((\d{4})\)', expand=False)
movies_df['title'] = movies_df['title'].str.replace(r'\(\d{4}\)', '', regex=True)
movies_df['title'] = movies_df['title'].str.strip()
ratings_df = ratings_df.drop('timestamp', axis=1)
```

```python
# داده‌های ژانرها
movies_df['genres'] = movies_df['genres'].str.split('|')

# فیلم‌ها با ژانرهای One-Hot Encoding
moviesWithGenres_df = movies_df.copy()
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        moviesWithGenres_df.at[index, genre] = 1
moviesWithGenres_df = moviesWithGenres_df.fillna(0)
```

```python
# تابع جستجو فیلم در دیتابیس
def search_movie(movie_name):
    movie = movies_df[movies_df['title'].str.contains(movie_name, case=False, na=False)]
    if not movie.empty:
        return movie[['title', 'year', 'genres']]
    else:
        return "فیلمی با این نام پیدا نشد."

# تابع فیلم‌های مشابه بر اساس ژانر
def similar_movies(movie_name):
    movie = movies_df[movies_df['title'].str.contains(movie_name, case=False, na=False)]
    if movie.empty:
        return "فیلمی با این نام پیدا نشد."

    movie_genres = movie.iloc[0]['genres']
    similar = movies_df[movies_df['genres'].apply(lambda x: bool(set(x) & set(movie_genres)))]
    return similar[['title', 'genres']]
```

```python

# تابع توصیه‌گر بر اساس ضریب همبستگی پیرسون
def get_movie_recommendations(user_input):
    inputMovies = pd.DataFrame(user_input)

    # اضافه کردن آیدی فیلم‌ها به داده‌های ورودی
    inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
    inputMovies = pd.merge(inputId, inputMovies)
    inputMovies = inputMovies.drop('year', axis=1)

    # زیرمجموعه‌ای از کاربران که فیلم‌های مشابه با فیلم‌های ورودی را مشاهده کرده‌اند
    userSubset = ratings_df[ratings_df['movieId'].isin(inputMovies['movieId'].tolist())]

    # گروه‌بندی اعضا بر اساس userId
    userSubsetGroup = userSubset.groupby(['userId'])
    userSubsetGroup = sorted(userSubsetGroup, key=lambda x: len(x[1]), reverse=True)
    userSubsetGroup = userSubsetGroup[0:100]

    # محاسبه ضریب همبستگی پیرسون
    pearsonCorrelationDict = {}
    for name, group in userSubsetGroup:
        group = group.sort_values(by='movieId')
        inputMovies = inputMovies.sort_values(by='movieId')
        nRatings = len(group)
        temp_df = inputMovies[inputMovies['movieId'].isin(group['movieId'].tolist())]
        tempRatingList = temp_df['rating'].tolist()
        tempGroupList = group['rating'].tolist()
        Sxx = sum([i**2 for i in tempRatingList]) - pow(sum(tempRatingList), 2)/float(nRatings)
        Syy = sum([i**2 for i in tempGroupList]) - pow(sum(tempGroupList), 2)/float(nRatings)
        Sxy = sum(i * j for i, j in zip(tempRatingList, tempGroupList)) - sum(tempRatingList) * sum(tempGroupList)/float(nRatings)
        if Sxx != 0 and Syy != 0:
            pearsonCorrelationDict[name] = Sxy / sqrt(Sxx * Syy)
        else:
            pearsonCorrelationDict[name] = 0

    pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
    pearsonDF.columns = ['similarityIndex']
    pearsonDF['userId'] = pearsonDF.index.map(lambda x: x[0] if isinstance(x, tuple) else x)
    pearsonDF.index = range(len(pearsonDF))

    # لیست کردن کاربران بر اساس ضریب همبستگی پیرسون
    topUsers = pearsonDF.sort_values(by='similarityIndex', ascending=False)[0:50]

    # ایجاد دیتافریم امتیازات وزنی
    topUsers['userId'] = topUsers['userId'].astype(int)
    topUsersRating = topUsers.merge(ratings_df, left_on='userId', right_on='userId', how='inner')
    topUsersRating['weightedRating'] = topUsersRating['similarityIndex'] * topUsersRating['rating']

    tempTopUsersRating = topUsersRating.groupby('movieId').sum()[['similarityIndex', 'weightedRating']]
    tempTopUsersRating.columns = ['sum_similarityIndex', 'sum_weightedRating']

    # محاسبه امتیاز پیشنهادی
    recommendation_df = pd.DataFrame()
    recommendation_df['weighted average recommendation score'] = tempTopUsersRating['sum_weightedRating'] / tempTopUsersRating['sum_similarityIndex']
    recommendation_df['movieId'] = tempTopUsersRating.index
    recommendation_df = recommendation_df.sort_values(by='weighted average recommendation score', ascending=False)
    recommendation_df = recommendation_df.reset_index(drop=True)

    # جستجوی نام فیلم‌ها با استفاده از movieId
    recommendation_df = recommendation_df.merge(movies_df[['movieId', 'title']], on='movieId', how='left')

    return recommendation_df[['title', 'weighted average recommendation score']].head(10)
```

```python

# تابع توصیه‌گر بر اساس محتوای فیلم
def get_content_based_recommendations(user_input):
    inputMovies = pd.DataFrame(user_input)

    # اضافه کردن آیدی فیلم‌ها به داده‌های ورودی
    inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
    inputMovies = pd.merge(inputId, inputMovies)
    inputMovies = inputMovies.drop(['genres', 'year'], axis=1)

      # استخراج ویژگی‌های ژانری برای فیلم‌های ورودی
    userMovies = moviesWithGenres_df[moviesWithGenres_df['movieId'].isin(inputMovies['movieId'].tolist())]
    userMovies = userMovies.reset_index(drop=True)
    userGenreTable = userMovies.drop('movieId', axis=1).drop('title', axis=1).drop('genres', axis=1).drop('year', axis=1)

    # محاسبه پروفایل کاربر بر اساس امتیازات داده شده به فیلم‌ها
    userProfile = userGenreTable.transpose().dot(inputMovies['rating'])

    # استخراج جدول ژانرها از دیتافریم اصلی
    genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])
    genreTable = genreTable.drop('movieId', axis=1).drop('title', axis=1).drop('genres', axis=1).drop('year', axis=1)

    # محاسبه امتیاز پیشنهادی برای هر فیلم بر اساس پروفایل کاربر
    recommendationTable_df = ((genreTable * userProfile).sum(axis=1)) / (userProfile.sum())

    # مرتب‌سازی نتایج و پیشنهاد 20 فیلم برتر
    recommendationTable_df = recommendationTable_df.sort_values(ascending=False)

    # نمایش نتایج
    recommended_movies = movies_df.loc[movies_df['movieId'].isin(recommendationTable_df.head(20).keys())]

    return recommended_movies[['title', 'movieId']]
```

```python
# تابع توصیه‌گر برای انتخاب سیستم توصیه‌گر
def recommendation_system(user_input, system_choice):
    if system_choice == '1':
        return get_movie_recommendations(user_input)
    elif system_choice == '2':
        return get_content_based_recommendations(user_input)
    else:
        return "انتخاب نامعتبر است."

# ورودی از کاربر برای انتخاب گزینه‌ها
def user_interface():
    print("لطفاً یکی از گزینه‌ها را انتخاب کنید:")
    print("1. جستجو فیلم در دیتابیس")
    print("2. فیلم‌های مشابه")
    print("3. سیستم توصیه‌گر")
    choice = input("شماره گزینه خود را وارد کنید: ")

    if choice == '1':
        movie_name = input("نام فیلم را وارد کنید: ")
        print(search_movie(movie_name))
    elif choice == '2':
        movie_name = input("نام فیلم را وارد کنید: ")
        print(similar_movies(movie_name))  # این خط ایرادی ندارد
    elif choice == '3':
        # دریافت ورودی از کاربر برای فیلم‌ها و امتیازها
        def get_user_input():
            user_input = []
            while True:
                movie_title = input("نام فیلم را وارد کنید (برای خاتمه وارد 'done' را بزنید): ")
                if movie_title.lower() == 'done':
                    break
                try:
                    rating = float(input(f"امتیاز خود را برای فیلم '{movie_title}' وارد کنید: "))
                    user_input.append({'title': movie_title, 'rating': rating})
                except ValueError:
                    print("لطفاً یک امتیاز معتبر وارد کنید.")
            return user_input

        # دریافت ورودی از کاربر
        user_input = get_user_input()

        # انتخاب سیستم توصیه‌گر از سوی کاربر
        system_choice = input("1. توصیه‌گر بر اساس همبستگی پیرسون یا 2. توصیه‌گر بر اساس محتوای فیلم؟ (1 یا 2): ")
        if system_choice == '1':
            print(get_movie_recommendations(user_input))
        elif system_choice == '2':
            print(get_content_based_recommendations(user_input))
        else:
            print("انتخاب معتبر نیست.")

user_interface()
```
